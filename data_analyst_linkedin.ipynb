# Data Analyst Job listing in LinkedIn Analysis

## import packages

```python
import pandas as pd
import matplotlib.pyplot as plt
from ast import literal_eval
```

## Read CSV file and assign to dataframe

```python
# I used Google sheets to explore the original africa dataset and found that it contains 434 rows (17.8% of total) with Arabic language in criteria column.
# so I decided to remove all arabic rows before uploading clean data into the workspace.

df_usa = pd.read_csv("/kaggle/input/linkedin-data-analyst-jobs-listings/linkedin-jobs-usa.csv", converters = {'criteria': literal_eval})
df_can = pd.read_csv("/kaggle/input/linkedin-data-analyst-jobs-listings/linkedin-jobs-canada.csv", converters = {'criteria': literal_eval})
df_afr = pd.read_csv("/kaggle/input/linkedin-jobs-africa-cleaned/linkedin-jobs-africa_clean.csv", converters = {'criteria': literal_eval})
```

```python
# show example of each dataframe

display(df_usa.head())

display(df_can.head())

display(df_afr.head())
```

## Add country column to define which data related to which country

```python
# remark country for each file

df_usa['country'] = 'USA'
df_can['country'] = 'Canada'
df_afr['country'] = 'Africa'
```

## Combine into one dataframe

```python
# combine all dataframes into one dataframe

df = df_usa.append(df_can, ignore_index = True).append(df_afr, ignore_index = True)
```

```python
# check information

df.info()
```

#### *Only salary column has null values*

```python
# check proportion of null values in salary column

df['salary'].isnull().mean() * 100
```

### 88% of data-analyst jobs in USA, Canada, and Africa posted in LinkedIn did not provide any salary information
#### I will drop the column since it might not show meaningful insights for analysis

```python
df = df.drop('salary', axis = 1)
```

```python
# change posted_date column data type to datetime

df['posted_date'] = pd.to_datetime(df['posted_date'])
df.info()
```

## Expand criteria column to criteria dataframe

```python
df_criteria = pd.DataFrame([{k:v for d in row for k,v in d.items()} for row in df['criteria']])
```

```python
df_criteria.head()
```

## Merge criteria dataframe back 

```python
df = pd.concat([df.drop(['criteria'], axis=1), df_criteria], axis=1)
```

```python
df.head()
```

## Check Job function and Industries tag

### Check Job functions

```python
# check job functions distinct values
df['Job function'].unique()
```

```python
# some job functions tags are combine in single string which make me unable to count the actual number of each tag. So, I will separate the tags.

job_func = df['Job function'].str.replace(',? and ', ', ', regex = True).str.split(', ')

# check

job_func
```

```python
# create function to summarize number of tags

def summary(column):
    result = {}
    
    for item in column:
        try:
            for index in range(len(item)):
                if item[index] in result:
                    result[item[index]] += 1
                else:
                    result[item[index]] = 1
        except:
            pass
    
    return result
```

```python
sum_job_func = summary(job_func)
```

```python
# create job functions dataframe
df_job_functions = pd.DataFrame(sum_job_func.items(), columns = ['Job function', 'count']).sort_values(by = 'count', ascending = False)
```

### Create chart to show top 10 job functions for data analyst in USA, Canada, Africa

```python
function_plot = df_job_functions.head(10).sort_values('count').plot(kind = 'barh', x = 'Job function', color = 'salmon')
function_plot.set_title("Top 10 Job functions for data analyst")
```

### Check Industries

```python
# check industries distinct values
df['Industries'].unique()
```

```python
# some industries tags are combine in single string which make me unable to count the actual number of each tag. So, I will separate the tags.

indust = df['Industries'].str.replace(', and ', ', ', regex = True).str.split(', ')

# check

indust
```

```python
sum_industries = summary(indust)
```

```python
# create industries dataframe

df_industries = pd.DataFrame(sum_industries.items(), columns = ['Industries', 'count']).sort_values('count', ascending = False)
```

### Create chart to show top 10 Industries for data analyst in USA, Canada, Africa

```python
industries_plot = df_industries.head(10).sort_values('count').plot(kind = 'barh', x = 'Industries', figsize = (5,5), color = 'steelblue')
industries_plot.set_title("Top 10 Industries for data analyst")
```

### Create line chart to show trend of data-analyst jobs posted in LinkedIn

```python
trend_plot = df.groupby('posted_date')['title'].count().plot(kind = 'line', color = 'lightseagreen')
trend_plot.set_title("How data-analyst jobs posted in LinkedIn rise through year")
trend_plot.set_xlabel("Date")
```
